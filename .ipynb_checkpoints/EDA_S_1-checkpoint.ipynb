**Step - 5**

**Add a new column on existing data frame**

```python
# Create a dataframe with name, age, and city.
```

```python
names = ['sherlock', 'alice', 'asta', 'yuno']
age = [25, 19, 15, 16]
city = ['Japan', 'Japan', 'Japan', 'Japan']
data = zip(names, age, city)
inx = ['A','B','C','D']
cols = ['NAME', 'AGE', 'CITY']
df1 = pd.DataFrame(data, index = inx, columns = cols)
df1
```

- If you want to add a new column

- First check how many rows are there

- Example: In above dataframe we have 4 rows

- your new column also should have 4 rows

- suppose I want to create a new column: **Salary**

- make a list with 4 values

- so salary = [10000, 20000, 30000, 40000]

**SYNTAX**

- dataframe[column] = list of values

- salary = [10000, 20000, 30000, 40000]

- df['salary'] = salary

```python
salary = [10000, 20000, 30000, 40000]
df1['SALARY'] = salary
df1
```

**use case - 1**

- Create empty data frame first

- Create 3 list using list comprehension

- list1 = a value from 1 to 10

- list2 = square of 1 to 10

- list3 = cube of 1 to 10

- Finally make a dataframe

```python
import pandas as pd
```

```python
df_2 = pd.DataFrame()
df_2
```

```python
list1 = [i for i in range(1,11)]
list2 = [i**2 for i in range(1,11)]
list3 = [i**3 for i in range(1,11)]
df_2['list_val'] = list1
df_2['list_sqr'] = list2
df_2['list_cube'] = list3
df_2
```

**Drop the columns**

```python
# How to check the columns
df_2.columns
```

- If you want to drop a specific column

- we will use drop method

- we already know about difference between keywords and methods

- keywords will use directly 

- Methods will use with corresponding dataframe

- **Method_name:drop**

**df_2.drop()**

- In order to drop a column

- Column: which column you want to drop

- axis: Represents the speciified value is a row or column

    - axis = 0 means rows
 
    - axis = 1 means column
 
    - by default axis = 0
 
- inplace: will provide overwrite the dataframe or not

      - inplce = True, this indicates overwrite the dataframe

      - inplace = False, this indicates the modified dataframe will be use when you provide new variable for the operation

```python
df_2.drop('X4')

# we are not providing axis
# it assuming the provided value is a row
# by default 
```

```python
df_2.drop('X4',axis=1)

# after this in a new cell type df_2
# you will see original dataframe
# because we did not provided inplace

# if you want to use below ouput
# you need to assign a variable

#df_new = df_2.drop('X4',axis=1)
```

```python
df_2.drop('X4', axis=1, inplace=True)
```
